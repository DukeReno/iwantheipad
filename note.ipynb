```python
class Prize:
    def __init__(self, category, year):
        self.category = category
        self.year = year

    def to_dict(self):
        return {"category": self.category,
                "years": self.year}

        

class Nobelist:
    def __init__(self, name, country, prize):
        self.name = name
        self.country = country
        self.prize = prize

    def to_dict(self):
        dict = {
                "country": self.country,
                "prize": { 
                    }
                }
        dict["prize"] = self.prize.to_dict() 
        return dict

```

```python
file = open("nobelists","r")
lines = file.readlines()
nobelists = []
    
country = ""
for line in lines:
    splitted = line.split(",")
    
    if len(splitted) == 1:
        
        country = line.strip("\n")
    else:
        if len(splitted) == 3:
            nobelists.append(Nobelist(splitted[0],country,Prize(splitted[1],splitted[2].strip("\n"))))
        elif len(splitted) == 4:
            nobelists.append(Nobelist(splitted[0],country,Prize(splitted[2],splitted[3].strip("\n"))))
        
```

```python
print(len(lines))
print(len(nobelists))
```

```python
nobelist_dict = {}

for nobel in nobelists:
    print("Name: " + str(nobel.name) + " From: " + str(nobel.country) + " prize: " + str(nobel.prize.category) + " year: " + str(nobel.prize.year))
    nobelist_dict.update({nobel.name : nobel.to_dict()})
```

```python
import json
with open("database.json","wt") as data:
    data.write(json.dumps(nobelist_dict))
```
