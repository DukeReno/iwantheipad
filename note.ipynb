```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
import requests
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_Nobel_laureates")[0]
print(f'Total tables: {len(df)}')
print(type(df))
df.columns
```

```python
df.drop(df.tail(1).index, inplace=True)
```

```python
dict = { 'Physics':[], 'Chemistry':[], 'Physiology or Medicine':[], 'Literature':[],
       'Peace':[], 'Economics (The Sveriges Riksbank Prize)[13][a]':[]
}
    
```

```python
print(type(df))
for (id, row) in df["Physics"].items():
    for name in str(row).split("; "):
        dict["Physics"].append(name)

for (id, row) in df["Chemistry"].items():
    for name in str(row).split("; "):
        dict["Chemistry"].append(name)

for (id, row) in df["Physiology or Medicine"].items():
    for name in str(row).split("; "):
        dict["Physiology or Medicine"].append(name)

for (id, row) in df["Literature"].items():
    for name in str(row).split("; "):
        dict["Literature"].append(name)

for (id, row) in df["Peace"].items():
    for name in str(row).split("; "):
        dict["Peace"].append(name)
        
for (id, row) in df["Economics (The Sveriges Riksbank Prize)[13][a]"].items():
    for name in str(row).split("; "):
        dict["Economics (The Sveriges Riksbank Prize)[13][a]"].append(name)
        
```

```python

URL = "https://en.m.wikipedia.org/wiki/List_of_Nobel_laureates"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
look = ""
physics_links = []
physics_names = []

for i in range(0,len(dict["Physics"])):   
    look = soup.find("a", string=dict["Physics"][i], href=True)
    try:
        if "/wiki/" in look["href"]:
            physics_links.append(requests.get("https://en.m.wikipedia.org" + look["href"]))
            physics_names.append(dict["Physics"][i])
    except:
        continue

chemistry_links = []
chemistry_names = []
print("chems")
for i in range(0,len(dict["Chemistry"])):   
    look = soup.find("a", string=dict["Chemistry"][i], href=True)
    try:
        if "/wiki/" in look["href"]:
            chemistry_links.append(requests.get("https://en.m.wikipedia.org" + look["href"]))
            chemistry_names.append(dict["Chemistry"][i])
    except:
        continue

phys_links = []
phys_names = []
print("physiol")
for i in range(0,len(dict["Physiology or Medicine"])):   
    look = soup.find("a", string=dict["Physiology or Medicine"][i], href=True)
    try:
        if "/wiki/" in look["href"]:
            phys_links.append(requests.get("https://en.m.wikipedia.org" + look["href"]))
            phys_names.append(dict["Physiology or Medicine"][i])
    except:
        continue

literature_links = []
literature_names = []
print("literature")
for i in range(0,len(dict["Literature"])):   
    look = soup.find("a", string=dict["Literature"][i], href=True)
    try:
        if "/wiki/" in look["href"]:
            literature_links.append(requests.get("https://en.m.wikipedia.org" + look["href"]))
            literature_names.append(dict["Literature"][i])
    except:
        continue
peace_links = []
peace_names = []
print("peace")
for i in range(0,len(dict["Peace"])):   
    look = soup.find("a", string=dict["Peace"][i], href=True)
    try:
        if "/wiki/" in look["href"]:
            peace_links.append(requests.get("https://en.m.wikipedia.org" + look["href"]))
            peace_names.append(dict["Peace"][i])
    except:
        continue
    
economics_links = []
economics_names = []
print("econ")
for i in range(0,len(dict["Economics (The Sveriges Riksbank Prize)[13][a]"])):   
    look = soup.find("a", string=dict["Economics (The Sveriges Riksbank Prize)[13][a]"][i], href=True)
    try:
        if "/wiki/" in look["href"]:
            economics_links.append(requests.get("https://en.m.wikipedia.org" + look["href"]))
            economics_names.append(dict["Economics (The Sveriges Riksbank Prize)[13][a]"][i])
    except:
        continue
```
